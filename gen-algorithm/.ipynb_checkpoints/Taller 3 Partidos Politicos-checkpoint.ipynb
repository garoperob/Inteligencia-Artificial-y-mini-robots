programa político. Sin embargo, en el congreso conformado por 5 partidos, no es fácil su tránsito,
por lo que debe repartir el poder, conformado por ministerios y otras agencias del gobierno, con
base en la representación de cada partido. Cada entidad estatal tiene un peso de poder, que es el
que se debe distribuir. Suponga que hay 50 curules, distribuya aleatoriamente, con una distribución
no informe entre los 5 partidos esas curules. Defina una lista de 50 entidades y asígneles

```python
import numpy as np  # Importamos la biblioteca numpy para trabajar con arreglos y funciones matemáticas

# Fijar la semilla para reproducibilidad
np.random.seed(3)  # Establecemos la semilla para los números aleatorios, garantizando que los resultados sean reproducibles

# Distribución aleatoria de curules entre 5 partidos
curules = np.random.choice(range(1, 50), 5, replace=False)  # Seleccionamos aleatoriamente 4 números del 1 al 49 sin reemplazo, representando los curules iniciales

curules = np.sort(curules)  # Ordenamos los curules seleccionados de menor a mayor

# Calculamos la diferencia entre los elementos consecutivos de los curules para obtener la distribución
curules_distribution = np.diff(curules, prepend=0)  # Calculamos las diferencias entre los elementos consecutivos y añadimos un 0 al inicio para obtener las diferencias correctas desde el primer elemento

# Añadimos la cantidad restante de curules que no fueron seleccionados
curules_distribution[-2] = curules_distribution[-2]+ (50-curules_distribution.sum())  # Calculamos los curules restantes para completar los 50 y los añadimos al final de la distribución

# Imprimimos la distribución de curules entre los partidos
print("Distribución de curules entre los partidos:", curules_distribution)  # Mostramos en pantalla la distribución final de curules entre los 5 partidos
```

```python
# Crear 50 entidades y asignarles un peso político
entidades = ["Entidad " + str(i+1) for i in range(50)]  # Genera una lista de 50 entidades nombradas como 'Entidad 1', 'Entidad 2', ..., 'Entidad 50'

pesos_politicos = np.random.randint(1, 101, 50)  # Genera una lista de 50 pesos políticos aleatorios, cada uno entre 1 y 100 inclusive

entidades_pesos = list(zip(entidades, pesos_politicos))  # Combina las dos listas en una lista de tuplas, donde cada tupla contiene una entidad y su respectivo peso político

print("Entidades y sus pesos políticos:", entidades_pesos)  # Imprime la lista de entidades junto con sus pesos políticos


# Calcula el poder político total sumando los pesos de todas las entidades
poder_politico_total = sum(pesos_politicos)

# Imprime el poder político total
print("Poder político total:", poder_politico_total)
```

```python
import numpy as np  # Importamos numpy para trabajar con arreglos y funciones matemáticas
import pandas as pd  # Importamos pandas para trabajar con DataFrames y visualizar datos
from IPython.display import HTML, display  # Importamos HTML y display para mostrar tablas estilizadas en Colab



# Crear una lista de nombres de partidos
partidos = ["Partido " + str(i+1) for i in range(5)]  # Genera una lista de 5 partidos nombrados como 'Partido 1', 'Partido 2', ..., 'Partido 5'

def Generar_matriz(partidos,curules_distribution,entidades_pesos):
  # Crear una matriz de poder de 5x50 inicializada en ceros
  matriz_poder = np.zeros((5, 50))  # Crea una matriz de 5 filas y 50 columnas llena de ceros, donde se almacenarán los pesos políticos asignados a cada partido

  # Mantener un registro de las entidades asignadas
  entidades_disponibles = list(range(50))  # Lista de entidades disponibles para asignación

  # Asignar entidades a los partidos según la distribución de curules
  for i, curul in enumerate(curules_distribution):  # Itera sobre la distribución de curules, obteniendo el índice y la cantidad de curules para cada partido
      entidades_asignadas = np.random.choice(entidades_disponibles, curul, replace=False)  # Selecciona aleatoriamente el número de entidades igual a los curules para el partido actual, sin reemplazo
      for entidad in entidades_asignadas:  # Itera sobre las entidades asignadas
          matriz_poder[i, entidad] = pesos_politicos[entidad]  # Asigna el peso político de la entidad correspondiente en la matriz de poder para el partido actual
          entidades_disponibles.remove(entidad)  # Elimina la entidad asignada de la lista de disponibles

  # Crear un DataFrame con la matriz de poder para mejor visualización
  df_matriz_poder = pd.DataFrame(matriz_poder, index=partidos, columns=["Entidad " + str(i+1) for i in range(50)])

  # Agregar una columna de suma al final de cada fila
  Peso_total_partidos = df_matriz_poder.sum(axis=1)
  df_matriz_poder['Peso_total_partido'] = Peso_total_partidos
  return df_matriz_poder,Peso_total_partidos

#POBLACIONES/MATRICES DE PODER (10 CANTIDAD)

matriz_poder_1,Peso_total_partido_1 = Generar_matriz(partidos,curules_distribution,entidades_pesos)
matriz_poder_2,Peso_total_partido_2 = Generar_matriz(partidos,curules_distribution,entidades_pesos)
matriz_poder_3,Peso_total_partido_3 = Generar_matriz(partidos,curules_distribution,entidades_pesos)
matriz_poder_4,Peso_total_partido_4 = Generar_matriz(partidos,curules_distribution,entidades_pesos)
matriz_poder_5,Peso_total_partido_5 = Generar_matriz(partidos,curules_distribution,entidades_pesos)
matriz_poder_6,Peso_total_partido_6 = Generar_matriz(partidos,curules_distribution,entidades_pesos)
matriz_poder_7,Peso_total_partido_7 = Generar_matriz(partidos,curules_distribution,entidades_pesos)
matriz_poder_8,Peso_total_partido_8 = Generar_matriz(partidos,curules_distribution,entidades_pesos)
matriz_poder_9,Peso_total_partido_9 = Generar_matriz(partidos,curules_distribution,entidades_pesos)
matriz_poder_10,Peso_total_partido_10 = Generar_matriz(partidos,curules_distribution,entidades_pesos)

Lista_peso_total_partidos = [Peso_total_partido_1,Peso_total_partido_2,Peso_total_partido_3,Peso_total_partido_4,Peso_total_partido_5,Peso_total_partido_6,Peso_total_partido_7,Peso_total_partido_8,Peso_total_partido_9,Peso_total_partido_10]

# Generar la tabla HTML con estilo para permitir el desplazamiento horizontal
html = matriz_poder_1.to_html(classes='table table-striped', border=0)
html = f"""
<div style="overflow-x:auto;">
    {html}
</div>
"""

# Generar la tabla HTML con estilo para permitir el desplazamiento horizontal
html2 = matriz_poder_2.to_html(classes='table table-striped', border=0)
html2 = f"""
<div style="overflow-x:auto;">
    {html2}
</div>
"""

# Mostrar la tabla HTML estilizada en Google Colab
display(HTML(html))

df_pesos_totales = pd.DataFrame({
    'Matriz_poder_1': Peso_total_partido_1,
    'Matriz_poder_2': Peso_total_partido_2,
    'Matriz_poder_3': Peso_total_partido_3,
    'Matriz_poder_4': Peso_total_partido_4,
    'Matriz_poder_5': Peso_total_partido_5,
    'Matriz_poder_6': Peso_total_partido_6,
    'Matriz_poder_7': Peso_total_partido_7,
    'Matriz_poder_8': Peso_total_partido_8,
    'Matriz_poder_9': Peso_total_partido_9,
    'Matriz_poder_10': Peso_total_partido_10
}, index=partidos)

# Generar la tabla HTML con estilo para permitir el desplazamiento horizontal
html_pesos_totales = df_pesos_totales.to_html(classes='table table-striped', border=0)
html_pesos_totales = f"""
<div style="overflow-x:auto;">
    {html_pesos_totales}
</div>
"""

# Mostrar la tabla HTML estilizada en Google Colab
display(HTML(html_pesos_totales))
```

El objetivo como jefe de gobierno es que se pueda controlar el poder político que tienen todos los demás partidos para que estos se vean de cierta manera estimulados a dar paso a los proyectos de leyes que se desee. Para esta actividad se definió que un poder equitativo a lo largo de los 5 partidos permitiría el objetivo anteriormente expuesto, por ende en la función de aptitud se considera el mejor valor como aquel que tenga menor divergencia respecto al promedio del poder total a lo largo de los partidos.


```python
def sigmoide(x):
    return 1 / (1 + np.exp(-10*x))


from re import L
def func_aptitud (Peso_total_partido):
  aptitud=np.zeros(5)
  for i in range(0, len(aptitud)):
    aptitud[i]=1-sigmoide(Peso_total_partido[i]/(Peso_total_partido.sum()/len(Peso_total_partido)))
  aptitud_matriz_poder = aptitud.sum()
  return aptitud_matriz_poder

def proceso_seleccion (Lista_peso_total_partidos):
  lista_aptitud=np.zeros(len(Lista_peso_total_partidos))
  lista_probabilidad_seleccion=np.zeros(len(Lista_peso_total_partidos))
  for i in range(len(lista_aptitud)):
    lista_aptitud[i]=func_aptitud(Lista_peso_total_partidos[i])
  aptitud_total_poblacion=lista_aptitud.sum()
  for i in range(len(lista_aptitud)):
    lista_probabilidad_seleccion[i]=lista_aptitud[i]/aptitud_total_poblacion*100
  return lista_probabilidad_seleccion

proceso_seleccion(Lista_peso_total_partidos)
```
