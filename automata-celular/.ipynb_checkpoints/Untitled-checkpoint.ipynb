# Punto 2

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Parmetros
grid_size = 50  # Tamao de la cuadrcula
p_fire = 0.8    # Probabilidad de que un rbol se queme si un vecino est en llamas
steps = 50      # Nmero de iteraciones

# Estados de las celdas
EMPTY = 0    # Espacio vaco o quemado
TREE = 1     # rbol
BURNING = 2  # rbol en llamas

# Crear la cuadrcula inicial
forest = np.random.choice([EMPTY, TREE], size=(grid_size, grid_size), p=[0.2, 0.8])

# Encendemos un fuego inicial en el centro
forest[grid_size // 2, grid_size // 2] = BURNING

# Configurar los colores de la simulacin
cmap = mcolors.ListedColormap(["black", "green", "red"])  # Negro: quemado, Verde: rbol, Rojo: ardiendo

# Funcin para actualizar la cuadrcula
def update_forest(forest):
    new_forest = forest.copy()
    for i in range(1, grid_size - 1):
        for j in range(1, grid_size - 1):
            if forest[i, j] == TREE:
                # Si un vecino est en llamas, el rbol puede prenderse
                neighbors = [forest[i-1, j], forest[i+1, j], forest[i, j-1], forest[i, j+1]]
                if BURNING in neighbors and np.random.rand() < p_fire:
                    new_forest[i, j] = BURNING
            elif forest[i, j] == BURNING:
                # Un rbol en llamas se convierte en quemado
                new_forest[i, j] = EMPTY
    return new_forest

# Simulacin y visualizacin
plt.figure(figsize=(6,6))
for step in range(steps):
    plt.imshow(forest, cmap=cmap, vmin=0, vmax=2)
    plt.title(f"Paso {step}")
    plt.pause(0.1)  # Pequea pausa para visualizar la evolucin
    forest = update_forest(forest)

plt.show()
```

# Punto 3

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d

# Coordenadas de drogueras en San Gil (Reemplzalas con ms datos reales)
coordenadas = np.array([
    (6.562259752404993, -73.13603045609159),
    (6.559010574606059, -73.13957940003495),
    (6.559148837922768, -73.13422119055183),
    (6.557005752203054, -73.13811807017592),
    (6.550576439772433, -73.13735261167832),
    (6.55541571490844, -73.12694933482476),
    (6.55530430950155, -73.13386507844814),
    (6.555215239137609, -73.13321186664567),
    (6.554655367914272, -73.13404012213377),
    (6.555071029943231, -73.13304109231824),
])

# Convertir lat/lon a un espacio 2D ms adecuado
# Escalamos latitud y longitud para evitar problemas de precisin
factor_escala = 100000  # Factor para evitar colinealidad
puntos = coordenadas * factor_escala  # Escala los datos

# Generar el diagrama de Voronoi
vor = Voronoi(puntos)

# Dibujar el diagrama
fig, ax = plt.subplots(figsize=(8, 8))
voronoi_plot_2d(vor, ax=ax, show_vertices=False, line_colors='blue', line_width=1)

# Marcar las drogueras en el mapa
ax.scatter(puntos[:, 1], puntos[:, 0], color='red', label="Drogueras")

# Configurar ttulo y leyenda
plt.title("Diagrama de Voronoi - Drogueras en San Gil")
plt.xlabel("Longitud (escalada)")
plt.ylabel("Latitud (escalada)")
plt.legend()
plt.grid()
plt.show()
```
